```r
nhanes <- read.csv("data/nhanes.csv")
```

```r
hist(nhanes$Height)
```

```r
library(moments)
```

```r
c(
  mean(nhanes$Height, na.rm=T), 
  median(nhanes$Height, na.rm=T), 
  sd(nhanes$Height, na.rm=T), 
  var(nhanes$Height, na.rm=T), 
  skewness(nhanes$Height, na.rm=T), 
  kurtosis(nhanes$Height, na.rm=T)
)
```
