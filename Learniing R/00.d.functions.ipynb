```r
bmi = function(weight, height){
  ans = weight/(height^2)
  status = NA 
  if(ans < 18.5){
    status = 'underweight'
  } else if(ans < 25){
    status = 'normal'
  } else if(ans < 30){
    status = 'overweight'
  } else {
    status  = 'obese'
  }
  return(list(bmi=ans, status=status))
}
```

```r
bmi(90, 1.7)
```

```r
calculate_bmi <- function(weight, height) {
  if (missing(weight) || missing(height)) {
    stop("Please provide both weight (kg) and height (m).")
  }
  if (!is.numeric(weight) || !is.numeric(height) || weight <= 0 || height <= 0) {
    stop("Weight and height must be positive numeric values.")
  }
  
  bmi <- weight / (height ^ 2)
  
  status <- if (bmi < 18.5) {
    "Underweight"
  } else if (bmi < 25) {
    "Normal"
  } else if (bmi < 30) {
    "Overweight"
  } else {
    "Obese"
  }
  
  return(list(BMI = round(bmi, 1), Status = status))
}
```

```r
calculate_bmi(90, 1.7)
```
